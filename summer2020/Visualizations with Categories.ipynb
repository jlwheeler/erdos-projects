```python
import numpy as np
import csv
import random
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
from palettable.colorbrewer.qualitative import Dark2_8
```

```python
spotify_meta = pd.read_csv('metadata_v12.csv')
```

```python
spotify_meta['First_category'].unique()
```

```python
for i in range(len(spotify_meta)):
    spotify_meta['First_category'][i] = spotify_meta['First_category'][i].replace(" ", "")
```

```python
for i in range(len(spotify_meta)):
    spotify_meta['First_category'][i] = spotify_meta['First_category'][i].replace("&", "")
```

```python
spotify_meta['First_category'].unique()
```

```python
text = " ".join(category for category in spotify_meta.First_category)
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(width = 700, height = 600, min_font_size = 6,
                      max_font_size = 80,
                      max_words=21, background_color="black", 
                      color_func=lambda *args, **kwargs: (30, 215, 97)).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

#To save the image:
wordcloud.to_file("category_cloud.png")
```
