```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import ast

#Convert these columns to list type
generic = lambda x: ast.literal_eval(x)
conv = {'Categories': generic,
        'SubCategories': generic,
        'show_name_emojis': generic,
        'show_desc_emojis': generic,
        'episode_desc_emojis': generic}

spotify_meta = pd.read_csv('metadata_v10.csv', converters=conv)
```

```python
spotify_meta.columns
```

```python
spotify_meta.head(1)
```

## Step 1: Clean data - Categories

```python
#Some categories are entered as an int 0
for i in range(len(spotify_meta)):
    if spotify_meta['Categories'][i] == 0:
        print(i, spotify_meta['Categories'][i])
```

```python
#Replace 0 ints with the list ['Unknown']

spotify_meta['Categories_clean'] = spotify_meta.loc[:,'Categories'].apply(lambda x: x if x != 0 else list(['Unknown']))
```

```python
for i in range(len(spotify_meta)):
    if 0 in spotify_meta['Categories_clean'][i]:
        spotify_meta['Categories_clean'][i].remove(0)
```

```python
#Labeling blank lists as ['Unknown']
spotify_meta['Categories_clean'] = spotify_meta.loc[:,'Categories_clean'].apply(lambda x: x[0] if len(x) != 0 else 'Unknown')
```

```python
#Some categories are entered as an int 0
for i in range(len(spotify_meta)):
    if spotify_meta['Categories_clean'][i] == 0:
        print(i, spotify_meta['Categories_clean'][i])
```

```python
spotify_meta.groupby('Categories_clean').count()
```

```python
spotify_meta.loc[:,'Categories_clean'] = spotify_meta.loc[:,'Categories_clean'].replace({'Sports': 'Sports & Recreation', 
                                                 'Games & Hobbies': 'Sports & Recreation', 
                                                 'Government': 'Government & Organizations', 
                                                 'Health': 'Health & Fitness'}) 
```

```python
spotify_meta.groupby('Categories_clean').count()
```

```python
sns.countplot(x="Categories_clean", data=spotify_meta)
plt.xticks(rotation=90)
```

## Step 2: Clean data - One Hot Encode

```python
#One hot encode explicit. 1 means yes it is explicit
spotify_meta['Explicit_encoded'] = pd.get_dummies(spotify_meta['Explicit'])['yes']
```

```python
#One hot encode episode description T/F emojis. 1 means it contains emojis
spotify_meta['ep_desc_emojis_encoded'] = pd.get_dummies(spotify_meta['episode_desc_contains_emojis'])[True]
```

## Some practice with random forest

```python
# Build a Decision Tree on this data
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

## First Attempt at a decision tree will predict whether the episode description contains emojis... :)

```python
# Use a max depth of 4 as a start
# Use 614 as a random state
# Call the classifier tree_clf
max_depth = 10
tree_clf = DecisionTreeClassifier(max_depth = max_depth,random_state = 614)
```

```python
from sklearn.model_selection import train_test_split
features = ['duration','ep_desc_lc','transcript_length','transcript_lc']
X_train,X_test,y_train,y_test = train_test_split(spotify_meta[features],
                                                spotify_meta[['episode_desc_contains_emojis']], 
                                                shuffle = True, 
                                                stratify = spotify_meta[['episode_desc_contains_emojis']],
                                                random_state = 614)
```

```python
# Fit the tree
tree_clf.fit(X_train,y_train)
```

```python
tree_clf.score(X_test,y_test)
```

```python
y_pred = tree_clf.predict(X_test)
```

```python
y_pred
```

```python
y = np.array(y_test['episode_desc_contains_emojis'])
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python

# Making function to calculate various stats
## Accuracy
def get_acc(prediction,y):
    return np.round(np.sum(prediction == y)/len(y),5)*100
## Precision
def get_prec(prediction,y):
    positives = prediction[prediction == 1]
    return np.round(np.sum(positives == y[prediction == 1])/len(positives),5)*100
## Recall
def get_recall(prediction,y):
    ones = y[y==1]
    return np.round(np.sum(prediction[y==1] == ones)/len(ones),5)*100
```

```python
get_acc(y_pred, y)
```

```python
get_prec(y_pred, y)
```

```python
get_recall(y_pred, y)
```
